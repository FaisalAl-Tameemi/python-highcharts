```python
import charts
```

```python
aapl = charts.data.aapl()
msft = charts.data.msft()
ohlc = charts.data.ohlc()

series = [
    aapl,
    msft,
    ohlc
]
options = charts.options.default()
```

```python
options['title'] = dict(text='Asynchronous chart')
stock = charts.plotasync(
    [aapl], options, height=500, stock=True, show='inline', 
    purge=True, display=False)

stock.plot(ohlc)
stock.plot(dict(data=[3,2,2], name='C'))
stock.plot(dict(data=[3,2,2], name='A'))

stock.show()
```

```python
stock.plot(ohlc)
```

```python
stock.plot(msft['data'], 'MSFT')
```

```python
chart = charts.plotasync(
    dict(data=[1,2,1], name='A'), options, height=500, show='tab', purge=False, save='temp')
chart.show()
```
